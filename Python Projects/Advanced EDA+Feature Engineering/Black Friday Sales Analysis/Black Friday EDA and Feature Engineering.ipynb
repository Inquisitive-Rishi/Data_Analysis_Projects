## Providing categorical encoding to Gender Column

```python
df["Gender"] = df['Gender'].map({"F":0,"M":1})
```

```python
df
```

## Providing categorical encoding to Age Column

```python
df["Age"].unique()
```

```python
df["Age"] = df["Age"].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7 })
```

```python
df.head()
```

## Providing Categorical Encoding to City_Category Column

```python
df["City_Category"].value_counts()
```

```python
city_df = pd.get_dummies(df["City_Category"],drop_first=True)
```

```python
df=pd.concat([df,city_df],axis=1)
```

```python
df.head()
```

```python
##Drop City_Category column
```

```python
df.drop(columns="City_Category",inplace=True)
```

```python
df.head()
```

For Product_Category_2

```python
df["Product_Category_3"] = df["Product_Category_3"].fillna(df["Product_Category_3"].mode()[0])
```

```python
df.head()
```

```python
df["Stay_In_Current_City_Years"] = df["Stay_In_Current_City_Years"].replace("4+", 4)
```
