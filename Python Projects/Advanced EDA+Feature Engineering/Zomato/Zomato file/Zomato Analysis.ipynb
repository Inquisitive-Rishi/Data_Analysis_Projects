```python
cc = pd.read_excel("Country-Code.xlsx")
cc
```

## Merging two different files

```python
merged = pd.merge(zomato,cc,on='Country Code',how='left')
merged
```

```python
country_count = merged["Country"].value_counts()
country_count
```

```python
country_name = country_count.index
country_name
```

```python
country_val = country_count.values
country_val
```

```python
plt.pie(country_val[:3],labels=country_name[:3],autopct="%1.2f%%")
plt.show()
```

Observation:
    > more than 90% transactions/business/sales are from India.
    > Followed by US then UK

```python
merged.columns
```

```python
rating_df = merged.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
rating_df
```

## Observations:
    1. Rating from 4.5 to 4.9 is --> Excellent
    2. Rating from 4.0 to 4.4 is --> Very Good
    3. Rating from 3.5 to 3.9 is --> Good
    4. Rating from 2.5 to 3.4 is --> Average
    5. Rating from 1.8 to 2.4 is --> Poor

```python
plt.figure(figsize=(9,5))
sns.barplot(x="Aggregate rating",y="Rating Count",data=rating_df)
plt.tight_layout()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',
            data=rating_df,
            palette=["white",'red','orange','yellow','green','green'])
plt.legend(loc="upper right")
```

## Observations:
    > Number of not rated are very high.
    > Majority of ratings are b/w 2.8 and 3.9

```python
sns.countplot(x='Rating color',palette=["blue",'red','orange','yellow','green','green'],data=rating_df)
```
