#### Observation: City Hotels have more cancellation rates compared to Resort Hotels

```python
city_hotels = hotels[hotels["hotel"] == "City Hotel"]
city_hotels["is_canceled"].value_counts(normalize=True)
```

## Mean of Average Daily Rates (adr) in both Hotels

```python
resort_hotels = resort_hotels.groupby('reservation_status_date')[["adr"]].mean()
city_hotels = city_hotels.groupby('reservation_status_date')[["adr"]].mean()
```

```python
plt.style.use('classic')
plt.figure(figsize=(20,8))
plt.title("Average Daily Rate in Resort and City Hotels",fontsize=25)
plt.plot(resort_hotels.index, resort_hotels['adr'], label = "Resort Hotel")
plt.plot(city_hotels.index, city_hotels['adr'], label = "City Hotel")
plt.legend(fontsize=18)
plt.show()
plt.tight_layout()
```

```python
hotels["month"] = hotels["reservation_status_date"].dt.month
plt.figure(figsize=(15,8))
x1 = sns.countplot(x="month", hue="is_canceled", data = hotels, palette ="tab10")
legend_labels,_ = x1. get_legend_handles_labels()
x1.legend(bbox_to_anchor=(1,1))
plt.title("Monthly Reservation Status",size=25)
plt.xlabel("month")
plt.ylabel("number of reservations")
plt.legend(["not cancelled", "cancelled"])
plt.show()
```

### Observation: High Ticket Prices are the main reason for ticket cancellation

```python
plt.style.use("ggplot")
plt.figure(figsize=(15,8))
plt.title('ADR per-month', fontsize = 30)
plt.bar('month','adr', data=hotels[hotels['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
cancelled = hotels[hotels['is_canceled'] == 1]
top_10 = cancelled['country'].value_counts()[:10]
plt.title('Top 10 countries with highest cancellation')
plt.pie(top_10, autopct='%1.2f%%',labels=top_10.index)
plt.show()
```
