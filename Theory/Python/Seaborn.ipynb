## .load_dataset()

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
taxi = sns.load_dataset("taxis")
taxi.head()
# Note: This data is imported from github directly and does not reside on computer
# Seaborn uses pandas dataframe to load
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",style="sex")
```

## Seaborn Line Plot

```python
flight = sns.load_dataset("flights")
flight
```

```python
sns.lineplot(data=flight,x="year",y="passengers")
# the light blue padding outside the line is ci i.e confidence index
# To deactivate ci, add ci=None
# By default, estimator is set to 'mean', we can change it to 'sum'
```

```python
# Same plotting can be done by
flight.groupby('year')["passengers"].mean().plot(kind="line")
```

```python
trips = sns.load_dataset("taxis",parse_dates=["pickup","dropoff"])
trips["hour"] = trips["pickup"].dt.hour
```

```python
sns.lineplot(data=trips,x="hour",y="fare",hue="payment",style="color",ci=None)
```

> Till now in Seaborn we plotted specific plots individually i.e lineplot,scatterplot
> These are axis-level plots, i.e these prints one plot at a time
> Now, we'll see plots method that are 'figure-level' plots 
  i.e they will print a figure,
      that may contain multiple subplots

```python
They can be broadly divided into three categories | family
    1.Relplot(relational)
    2.Displot(Distribution)
    3.Catplot(Categorical)
```

```python
1. Relplot
    using .relplot(),
    we can print two types of plot:
        1.Scatterplot
        2.Lineplot

2. Displot
    using .displot(),
    we can print four types of plot:
        1.Histogram i.e histplot
        2.Kdeplot i.e Kernel Density Estimator
        3.Ecdfplot
        4.Rugplot
        
3. Catplot
    using .catplot(),
    We can print six types of plot:
        1.Stripplot
        2.Swarmplot
        3.boxplot
        4.Violinplot
        5.pointplot
        6.barplot
```

## The relplot() method

```python
> By default, relplot() plots a scatterplot,
> we can change it to line, kind="line"
> As relplot is a figure-level method
```

```python
> We can create subplot by defining rows and columns
> This creates figure with subplots in it.
```

```python
sns.relplot(data=trips,x="hour",y="fare",
            kind="line",
            style="payment",
            ci=None,
            col="color"
            )
```

```python
sns.relplot(data=trips,x="hour",y="fare",
            kind="line",
            style="payment",
            ci=None,
            col="color",
            row="pickup_borough"
            )
```

## Resizing Seaborn plots arguments and height

```python
We can resize plots with:
    1.height
    2.aspect
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",
          row="sex",col="day",
          kind="scatter",
          height=3,aspect=1.5)
# Eacch facet height is 3 and width 1.5 times of height
```

## Seaborn Histogram

```python
sns.histplot(data=tips,x="tip",hue="smoker")
# Plots a layered-histogram with opening lowered
```

## KDE

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
sns.histplot(data=penguins,x="body_mass_g",hue="species",binwidth=200,kde=True)
```

## element="step"

```python
sns.histplot(data=penguins,x="body_mass_g",
             hue="species",
             element="step",
             multiple="stack"
            )
```

## KDE Plots

```python
sns.kdeplot(data=penguins,x="body_mass_g")
```

```python
sns.kdeplot(data=penguins,x="body_mass_g",hue="species")
```

## bw_adjust (Bandwidth Adjust)

decreasing bw_adjust inreases the accuracy of plot but increases the 'jaggedness'

```python
sns.kdeplot(data=penguins,x="body_mass_g",hue='species',bw_adjust=0.5)
```

## Bivariate Distribution Plot

```python
sns.histplot(data=penguins,x="body_mass_g",y="flipper_length_mm",
            hue="species"
            )
```

```python
sns.kdeplot(data=penguins,x="body_mass_g",y="flipper_length_mm",
            hue="species"
            )
```

## Rugplot

```python
plt.figure(figsize=(4,4))
sns.rugplot(data=tips,x="tip")
```

```python
plt.figure(figsize=(4,4))
sns.rugplot(data=tips,x="tip",height=1)
```

```python
plt.figure(figsize=(4,4))
sns.rugplot(data=tips,y="tip",height=1)
```

## We can plot two graphs in one

```python
plt.figure(figsize=(4,4))
sns.kdeplot(data=tips,x="total_bill")
sns.rugplot(data=tips,x="total_bill")
```

```python
plt.figure(figsize=(4,4))
sns.scatterplot(data=tips,x="total_bill",y="tip")
sns.rugplot(data=tips,x="total_bill",y="tip")
# Some data with tons of layers, we can change the opacity of rugplot
```

```python
diamond = sns.load_dataset("diamonds")
diamond.head()
```

```python
sns.scatterplot(data=diamond,x="carat",y="price",s=5)
sns.rugplot(data=diamond,x="carat",y="price",lw=1,alpha=0.005)
```

## Displot 

```python
sns.displot(kind="hist",data=penguins,x="body_mass_g",col="species",
           height=4,aspect=2)
```

```python
sns.displot(data=tips,kind="kde",x="tip",col="time",rug=True)
```

```python
sns.displot(data=tips,kind="kde",x="total_bill",y="tip",rug=True)
```
