# Seaborn part-2

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Countplot

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.countplot(data=penguins,x="species")
# To make bars horizontal, change axis from x to y
```

```python
# Same can be done using pandas
penguins["species"].value_counts().plot(kind="bar")
```

## Stripplot and Swarmplot

## Boxplot

```python
sns.boxplot(data=penguins,x="body_mass_g")
# Change the boxplot orientation to horizontal by changing x to y 
```

```python
sns.boxplot(data=penguins,y="body_mass_g")
```

```python
sns.boxplot(data=penguins,x="species",y="body_mass_g")
```

We can plot swarmplot with boxplot

```python
sns.boxplot(data=penguins,x="species",y="body_mass_g")
sns.swarmplot(data=penguins,x="species",y="body_mass_g",color="0.4")
```

```python
sns.boxplot(data=penguins,x="body_mass_g",whis=1)
```

```python
sns.boxplot(data=penguins,x="body_mass_g",whis=0.5)
```

## Boxenplot

Boxenplot is Boxplot on 'steroids'

```python
sns.boxenplot(data=penguins,x="body_mass_g")
```

## Violinplot

```python
titanic = pd.read_csv("titanic.csv")
titanic.head()
```

```python
titanic["age"] = titanic["age"].replace(["?"],np.nan)
titanic["age"] = titanic["age"].astype(float)
```

```python
sns.violinplot(data=titanic,x="age")
sns.boxplot(data=titanic,x="age")
```

```python
sns.violinplot(data=titanic,x="pclass",y="age")
```

We can split violnplots
First we separate violinplots on hue="sex"

```python
sns.violinplot(data=titanic,x="pclass",y="age",hue="sex")
```

but we can split them on the same graph

```python
sns.violinplot(data=titanic,x="pclass",y="age",hue="sex",split=True)
# Creates 3 Violinplot with half of each is sex and half of it is pclass
```

## Barplot

```python
taxi = sns.load_dataset("taxis")
taxi.head()
```

```python
sns.barplot(data=taxi,x="pickup_borough",y="fare")
# By default,Barplot calculate mean of numerical date on y-axis
```

```python
# We can change the 'mean' to 'sum'
sns.barplot(data=taxi,x="pickup_borough",y="fare",estimator=sum)
```

We can do the same thing in Pandas

```python
taxi.groupby("pickup_borough")["total"].sum().plot(kind="bar")
```

```python
taxi.groupby("pickup_borough")["total"].sum().plot(kind="bar")
```

```python
sns.barplot(data=taxi,y="pickup_borough",x="fare",hue="color",
            dodge=False,orient="h")
```

## The Catplot

```python
sns.catplot(data=titanic,x="sex",y="survived",
           kind="bar",col="pclass"
           )
```

## Swarmplot

```python
sns.catplot(data=taxi,kind="strip",x="pickup_borough",y="distance")
```
