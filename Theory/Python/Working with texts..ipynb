## Stripping whitespace using .strip()

```python
s = pd.Series(["1.   Arjun  ","  2. Drona\n  ","  3.  Bhishma "])
s
```

```python
s.str.strip()
```

```python
# We can also choose what to strip
s.str.strip(to_strip="123. \n")
```

```python
# Removes whitespace from the left
s.str.lstrip()
```

```python
# Remove whitespace from the right
s.str.rstrip()
```

## Splitting Text values with .split()

```python
titanic = pd.read_csv("titanic.csv")
titanic.head()
```

```python
titanic["home.dest"]
```

```python
titanic["home.dest"].str.split()
```

```python
# We can split these into separate rows i.e home, destination
# Here we will defina a separator.
# .split() by default separates on the basis of 'space'
titanic["home.dest"].str.split("/")
```

```python
# Now we will define separate columns for home and destination
titanic["home.dest"].str.split("/",expand=True)[0]
```

```python
titanic["home"] = titanic["home.dest"].str.split("/",expand=True)[0]
titanic["destination"] = titanic["home.dest"].str.split("/",expand=True)[1]
```

```python
titanic.head()
```

```python
# Here expand prints a dataframe
# We can define how many times to split by using n 
titanic["home.dest"].str.split("/",expand=True,n=1)
```

## Replacing portion of string with replace()

```python
# .replace() replaces string from all rows of a column.
```

```python
titanic["cabin"].str.replace("?","???")
```

```python
ufo = pd.read_csv("nuforc_reports.csv")
ufo["duration"].str.replace("seconds","sec")
```

```python
# To replace two or more strings, we use OR(\) regex
ufo["duration"].str.replace("seconds|minutes","",regex=True)
```

```python
# Dynamically apply strings by using custom function
def abbrev(reo):
    print(reo)
    return"HI"
# If we include this in the code, it prints all the rows with 'reo' 
# reo:'Regular Expression Object'
```

```python
ufo["duration"].str.replace("seconds|minutes",abbrev)
```

```python
# We can make custom function to extract only first string charater from string.
def abbrev(reo):
    return reo.group()[0]
# This prints only first character from string
```

```python
ufo['duration'].str.replace("seconds|minutes",abbrev)
```

## Testing strings with .contains()

```python
# .contains() checks if the columns contain a string in rows and returns boolean.
ufo["duration"].head()
```

```python
ufo["duration"].str.contains("s").head(10)
```

```python
# It shows error if NaN is present
ufo[ufo["duration"].str.contains("s")]
# To avoid this error we use na=False
```

```python
ufo[ufo["duration"].str.contains("s",na=False)].head()
```

```python
ufo[ufo["duration"].str.contains("day|week|month",na=False)].head()
```
