## Delete rows that are not numerically indexed

```python
countries
```

```python
countries.drop(countries.index[3:142])
```

## Creating new dynamic column

```python
titanic = pd.read_csv("titanic.csv")
titanic.head()
# We can make new column by adding multiple columns
```

```python
titanic["total_relatives"] = titanic["sibsp"] + titanic["parch"]
```

```python
titanic.head()
```
