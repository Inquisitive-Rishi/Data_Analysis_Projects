# Filtering Dataframes

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
bs = pd.read_csv("bestsellers.csv")
bs.head()
```

## The betweeen() method

```python
bs["User Rating"].between(3.5,3.8)
# Filters books with ratings b/w 3.5 and 3.8
```

```python
bs[bs["User Rating"].between(3.5,3.8)]
```

```python
df = bs[bs["User Rating"].between(3.5,3.8)]
df["User Rating"].value_counts()
```

## isin() method

filter different values from a column

```python
netflix = pd.read_csv("netflix_titles.csv",sep="|")
netflix[netflix["country"].isin(["India", "Japan", "South Korea"])]
```

```python
netflix[netflix["country"].isin(["India", "Japan", "South Korea"])]["country"].value_counts()
```

```python
# Filter shows based on rating
netflix[netflix["rating"].isin(["TV-MA","R"])]
```

## Combining condition using AND (&)


```python
houses = pd.read_csv("kc_house_data.csv")
```

```python
view = houses["view"] == 4
grade = houses["grade"] >= 10
```

```python
view & grade
```

```python
houses[view & grade]
```

```python
houses[(houses["view"] == 4) & (houses["grade"]>10)]
```

```python
d1 = netflix["director"].isin(['David Fincher',])
d2 = netflix["director"].isin(['Julien Leclercq',])
```

```python
netflix[d1|d2]["director"].value_counts()
```

## Bitwise Negation (~)

```python
titanic = pd.read_csv("titanic.csv")
```

```python
survived = titanic["survived"] == 1
```

```python
survived
```

```python
not_survived = ~survived
not_survived
```

## isna() and notna()

```python
netflix[netflix["director"].notna()]
# prints dataframe of netflix without all directors excluding all null values
```
