## Sorting by multiple columns

```python
# To sort a column
houses = pd.read_csv("kc_house_data.csv")
houses.sort_values("bedrooms")
```

```python
houses.sort_values(["bedrooms","bathrooms"],ascending=False)
# First bedrooms gets sorted then bathrooms gets sorted.
```

```python
> Sorting by number is simple.
but, when sorting is done by letter/alphabets

> Few points to remember:
    sort_values() sorts this way.
    ABCD...Zabcd....z
    
    A < a
    c > E
    Uppercase < Lowercase
```

```python
titanic = pd.read_csv("titanic.csv")
titanic.sort_values("name",key=lambda col:col.str.lower())
```

```python
titanic.sort_values("name",key=lambda col:col.str.lower())
```

## .loc[]

```python
# To print dataframes based on rows.
countries.loc["India"]
```

```python
countries.loc[["China","United States","India"]]
```

```python
countries.sort_index(inplace=True)
countries.loc["China":"India"]
```

```python
titanic.loc[12:50:6]
```

## .iloc[ ]

```python
> While .loc[] prints "named index" or 'index'.

> .iloc prints based on position of the rows even if the dataframe is not indexed as numeric
```

```python
countries.iloc[[5,10,15]]
```

## .loc[ ] and .iloc[ ] with series

```python
# Print specific rows and specific columns
countries.loc[["Afghanistan","Pakistan","India"],["Ladder score","Social support"]]
```

```python
titanic.loc[10:50:2,["name","age","embarked"]]
```

```python
# We can also write like this
countries.loc["Canada":"Denmark"]["Ladder score"]
```

```python
houses["bedrooms"].value_counts().iloc[0]
# Prints the total no:of bedroom counts on zero index.
```

```python
houses["bedrooms"].value_counts().loc[33]
# Prints the total no:of houses with 33 bedrooms
```

```python
titanic["age"].value_counts().loc["18"]
# Here 18 is string beause age was string type.
```
