## Rename a Column and Index

## replace() method

```python
pokemon = pd.read_csv("Pokemon.csv")[["Name","Type 1", "Type 2"]].head()
pokemon.head()
```

```python
pokemon["Type 1"].replace(["Fire","Grass"],["Aag","Ghaans"])
```

```python
pokemon["Type 1"].value_counts()
```

```python
pokemon["Type 2"].value_counts()
```

```python
pokemon["Type 2"].value_counts(dropna=False)
```

```python
pokemon["Type 2"].replace(np.nan,"IDK")
```

```python
# or, replace NaN with pandas 'None'
pokemon["Type 2"].replace([None],"IDK")
```

## Updating values using .loc[]

```python
countries.loc[["India","Afghanistan","Pakistan","Bangladesh"]]
```

```python
# To change regional indicator and index score of all four countries
countries.loc[["India","Afghanistan","Pakistan","Bangladesh"],["Regional indicator","Ladder score"]]=["Akhand Bharat","Avg"]
# This updates the value of both columns.
```

```python
countries.loc[["India","Afghanistan","Pakistan","Bangladesh"]]
```

## Making update with loc[] and Boolean Mask

```python
houses = pd.read_csv("kc_house_data.csv")
houses.head()
```

```python
# First we create a boolean mask.
good_quality = houses["grade"]>12
great_view = houses["view"] == 4
# We created two boolean masks
```

```python
houses[good_quality & great_view]
```

```python
houses["luxury_level"] = "Normal"
```

```python
# Now, we'll update these filtered datasets luxury_level to "Ultra"
houses.loc[good_quality & great_view,["Luxury_level"]] = "Ultra"
# This keeps all rows of luxury_levels to "Normal" but, updates all filtered rows to "Ultra"
```

```python
houses["Luxury_level"].value_counts()
```
