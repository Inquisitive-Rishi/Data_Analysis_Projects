# Matplotlib - 2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Barplot

```python
seeds = ["Wheat","Spinach","Tulip","Cabbage","Guava"]
germinated = [78,86,68,72,80]
died = [15,25,32,18,38]
```

```python
plt.bar(seeds,germinated)
# We can add many things to its bar graph hange the color,width etc
```

```python
plt.bar(seeds,germinated,color="limegreen",width=0.4)
plt.bar(seeds,died,color="red",width=0.5)
```

```python
# We can align position of overlapping bars so that each bars are fully visible.
plt.bar(seeds,germinated)
plt.bar(seeds,died,align="edge")
# here blue color is germinated, and orange color is died.
```

```python
# We can also change position of bars (top/bottom)
plt.bar(seeds,germinated)
plt.bar(seeds,died,bottom=germinated)
```

```python
# we can create horizontal bar
plt.barh(seeds,germinated)
```

## Creating Histogram

```python
# Histogram presents the frequency of items i.e how many time the items occured
nums = [1,2,1,1,2,2,3,4,5,5,4,3,6,7,8,7,5]
plt.hist(nums,bins=10)
```

```python
titanic = pd.read_csv("titanic.csv")
plt.hist(titanic["age"],bins=20)
plt.show()
```

```python
first_class = titanic[titanic["pclass"] == 1]["age"] 
second_class = titanic[titanic["pclass"] == 2]["age"]
```

```python
plt.hist(first_class,alpha=0.4)
plt.hist(second_class,alpha=0.4)
```

## Pie Chart

```python
lunch = ["Dosa","Pineapple pie","Mushrooms","Chicken","Omelette"]
price = [9.00,5.38,3.86,12.03,6.83]
```

```python
plt.pie(price,labels=lunch,autopct="%1.2f%%")
plt.show()
```

```python
# Using explode
pokemon = pd.read_csv("Pokemon.csv").head(4)
plt.pie(pokemon["Attack"],labels=pokemon["Name"],shadow=True,autopct="%1.2f%%",explode=(0,0.1,0,0.1))
plt.show()
# Explode must be equal to length of no:of rows
```

## Creating Subplot

```python
num = np.arange(4)
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(10,4))
plt.suptitle("My great Subplot")
plt.subplot(1,3,1)
plt.plot(num,num,c="teal")
plt.title("x")

plt.subplot(1,3,2)
plt.plot(num,num**2)
plt.title("x-squared")

plt.subplot(1,3,3)
plt.plot(num,num**3)
plt.title("x-cubed")

plt.tight_layout()
```

```python
first_class = titanic[titanic["pclass"]==1]["age"]
second_class = titanic[titanic["pclass"]==2]["age"]
third_class = titanic[titanic["pclass"]==3]["age"]
```

```python
plt.figure(figsize=(10,4))
plt.suptitle("Amazing Subplot")
ax=plt.subplot(1,3,1)
plt.hist(first_class)
plt.title("1st Class")

plt.subplot(1,3,2, sharey=ax)
plt.hist(second_class,color="purple")
plt.title("2nd Class")

plt.subplot(1,3,3, sharey=ax)
plt.hist(third_class,color="green")
plt.title("3rd Class")
plt.tight_layout()
```
