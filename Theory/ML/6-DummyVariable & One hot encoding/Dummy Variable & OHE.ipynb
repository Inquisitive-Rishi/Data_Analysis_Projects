```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("city_area_price.csv")
df
```

```python
dummies = pd.get_dummies(df["town"])
dummies
```

```python
merged_df = pd.concat([df,dummies],axis=1)
merged_df
```

```python
df2 = merged_df.drop(columns=["town",'Gurgaon'])
df2
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
x = df2.drop(columns="price")
x
```

```python
y = df2["price"]
y
```

```python
reg.fit(x,y)
```

## Predicting price of home in Pune

```python
reg.predict([[3500,0,1]])
```

## Predicting price of home in Gurgaon

```python
reg.predict([[3860,0,0]])
```

```python
reg.score(x,y)
```
